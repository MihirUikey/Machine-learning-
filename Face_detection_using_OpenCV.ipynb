#import modules


```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

#HAAR cascade file path


```python
face_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/Colab Notebooks/project files/haarcascade_frontalface_default.xml')
```

#Load Image

```python
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/project files/face.jpg")
#convert to rgb
img_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
#convert to gray scale image
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap="gray")
```

#Detect Faces

```python
faces = face_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=4, minSize=(30,30))
```

```python
len(faces)
```

```python
#importing cv2_imshow(was getting error wihtout this lib)
from google.colab.patches import cv2_imshow
#code for detecting faces in image
for (x, y, width, height) in faces:
    cv2.rectangle(image, (x, y), (x+width, y+height), (0, 255, 0), 2)

cv2_imshow(image)
cv2.waitKey(0)
```

Random Forest model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
wine = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Lab on ml/50_Startups.csv')
wine
```

correlation matrix

```python
correlation_matrix = wine.corr()
correlation_matrix
```

#iloc function


```python
X =wine.iloc[:,5:7].values #two independent variables
Y= wine.iloc[:,11:12].values #one dependent variables
X,Y
```

#Training the random forest model

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor(n_estimators=10,random_state=0,min_samples_leaf=3)
reg.fit(X,Y)
```

#Description of random forest regression model

```python
print('Number of estimators: ',reg.n_estimators)
print('Feature importances: ',reg.feature_importances_)
print('Max depth of trees: ',[estimator .tree_.max_depth for estimator in reg.estimators_])
```

Hyperparameter tuning for Random Forest

```python
reg = RandomForestRegressor(n_estimators=15,random_state=0,min_samples_leaf=5,min_samples_split=4,max_features=4)
reg.fit(X,Y)
```

```python
print('Number of estimators: ',reg.n_estimators)
print('Feature importances: ',reg.feature_importances_)
print('Max depth of trees: ',[estimator .tree_.max_depth for estimator in reg.estimators_])
print('Minimum samples split: ',reg.min_samples_split)
print('Maximum features: ',reg.max_features)
```

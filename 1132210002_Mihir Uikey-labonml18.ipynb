#**Importing libraries and dataset**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
fitness = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Lab on ml/fitness_class_2212.csv')
fitness
```

#**Using iloc function, Splitting dataset into train and test dataset**

```python
from sklearn.model_selection import train_test_split
X = fitness.iloc[:,1:2].values
Y = fitness.iloc[:,-1].values
X
X_train, X_test, Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=0)

#printing X_Train dataset
print(X_train)
```
